# Up in the Air 

## Mission:
Utilizing a Aviationstack's API to build a smart contract/dApp that identifies whether a flight has been delayed (i.e. flight delays over 90 minutes) and accordingly automates and interacts with the Blockchain to file a claim and returns flight delay compensation to the customer. 

![delay](airportdelays.jpg)

## Import Libraries

```python
import os 
import requests
from dotenv import load_dotenv
```

```python
# Load .env enviroment variables
load_dotenv("api_key.env")
```

```python
aviationstack_api = os.getenv("AVIATIONSTACK_API_KEY")
#print(aviationstack_api)
```

```python
# Filtering results by providing a minimum amount of minutes in arrival and departure delay. Example: 90 for at least 90 minutes of delay in arrival.
params = {
  'access_key': aviationstack_api, 'min_delay_arr': 90, 'min_delay_dep': 90
}

api_result = requests.get('http://api.aviationstack.com/v1/flights', params)

api_response = api_result.json()
```
